```python
data["cast"] = data ["cast"].replace(np.nan,"None")
data.tail()
```

```python
cast_length_list = []
for i in range(0,7787):
    if data.loc[i, 'cast'] == 'None':
        cast_length = 0
        cast_length_list.append(cast_length)
    else:
        cast_length = len(data.loc[i,'cast'].split(', '))
        cast_length_list.append(cast_length)
```

```python
data['number_of_cast'] = cast_length_list
data.tail()
```
